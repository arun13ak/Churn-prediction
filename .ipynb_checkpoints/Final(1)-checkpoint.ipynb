```python
data['Attrition']=data['Attrition'].apply(lambda x:1 if x=="Yes" else 0)
data['OverTime']=data['OverTime'].apply(lambda x:1 if x=="Yes" else 0)
```

```python
data=data.drop(['Over18','StandardHours'],axis=1)
```

```python
data=data.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus'],axis=1)
```

```python
X=data.drop(['Attrition'],axis=1)
Y=data['Attrition']
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
models=[]
models.append(('LR',LogisticRegression()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('KNC',KNeighborsClassifier()))
models.append(('DTC',DecisionTreeClassifier()))
models.append(('NB',GaussianNB()))
models.append(('SVM',SVC()))
```

```python
results=[]
names=[] 
for name,model in models: 
    kfold=KFold(n_splits=10)
    cv_results=cross_val_score(model,X,Y,cv=kfold)
    results.append(cv_results)
    names.append(name)
    print(name,cv_results.mean()*100.0)
```

```python
estimators=[]
model1=LogisticRegression()
model2=SVC()
model3=KNeighborsClassifier()
model4=LinearDiscriminantAnalysis()
estimators.append(('LR',model1))
estimators.append(('SVM',model2))
estimators.append(('KNC',model3))
estimators.append(('LDA',model4))
```

```python
ensemble=VotingClassifier(estimators)
```

```python
kfold=KFold(n_splits=30)
results=cross_val_score(ensemble,X,Y,cv=kfold)
```

```python
results
```

```python
results.mean()*100
```

```python
ensemble.fit(x_train,y_train)
```

```python
print(ensemble.score(x_train,y_train))
print(ensemble.score(x_test,y_test))
```
